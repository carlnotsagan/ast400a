```python
# load data and see which variables are available
m1_history = pd.read_csv('data/m1_history.data',sep=r'\s+',header=4)
m2_history = pd.read_csv('data/m2_history.data',sep=r'\s+',header=4)
list(m1_history)
#m1_history.head(10)
```

```python
m1_history_total_mass = m1_history['star_mass']
m1_history_log_R = m1_history['log_R']

m2_history_total_mass = m2_history['star_mass']
m2_history_log_R = m2_history['log_R']

m1_history_age = m1_history['age']
m1_history_mdot_rate = m1_history['lg_mtransfer_rate']
```

```python
## 1 result here
plt.title(r'Binary Star System - $M_{1}=10M_{\odot} - M_{2}=8.9M_{\odot} - P_{i}=2.2$ days')



plt.plot(m1_history_total_mass,m1_history_log_R,
         color='dodgerblue', label=r'Primary $M_1$')

plt.plot(m2_history_total_mass,m2_history_log_R,
         color='goldenrod',label=r'Secondary $M_2$')

plt.text(10.2, 0.6, '(A)', fontsize=12)

plt.text(10.2, 0.85, '(B)', fontsize=12)


plt.legend()
plt.xlabel(r'$M_{\rm{total}} \ (M_{\odot})$')
plt.ylabel(r'$\rm{log}~R \ (R_{\odot})$')
```

```python
## 1 result here
plt.title(r'Binary Star System - $M_{1}=10M_{\odot} - M_{2}=8.9M_{\odot} - P_{i}=2.2$ days')


plt.plot(m1_history_age,m1_history_mdot_rate,
         color='dodgerblue')


plt.text(1.62e7, -7, '(B) - RLOF', fontsize=10)
plt.text(1.85e7, -6.3, '(C) - End of rapid/TE achieved', fontsize=10)
plt.text(2.15e7, -7.3, '(D) - End of long lived ', fontsize=10)


plt.ylim(-8,-3)
plt.xlim(1.6e7,2.4e7)

plt.xlabel(r'Age (yr)')
plt.ylabel(r'$\rm{Mass \ Transfer \ Rate} \ (M_{\odot} \ \rm{yr}^{-1})$')
```

## b. - HR Diagram of Binary Star System

Using the same data, 

1. **Plot** an HR diagram for the primary and secondary and label them. 
2. **Label** - the phase start of ZAMS (A) to the filling of the Roche lobe of the primary (B) using plt.annotate or similar for the primary. 

```python
m1_history_log_Teff = m1_history['log_Teff']
m1_history_log_L = m1_history['log_L']

m2_history_log_Teff = m2_history['log_Teff']
m2_history_log_L = m2_history['log_L']
```

```python
plt.plot(m1_history_log_Teff,
           m1_history_log_L,
         color='dodgerblue', label=r'Primary $M_1$')

plt.plot(m2_history_log_Teff,
           m2_history_log_L,
         color='goldenrod',label=r'Secondary $M_2$')

plt.gca().invert_xaxis()


plt.text(4.42, 3.8, '(A)', fontsize=12)

plt.text(4.33, 4, '(B)', fontsize=12)


plt.legend()
plt.xlabel(r'$\rm{log}~T_{\rm{eff}} \ (K)$')
plt.ylabel(r'$\rm{log}~L \ (L_{\odot})$')

plt.show()
```
