```python
# load data and see which variables are available
tp_agb_data = pd.read_csv('3m_to_3DU_history.data',sep=r'\s+',header=4)
tp_agb_data.head(10)
#list(tp_agb_data)
```

```python
# load vars here
three_m_star_age_years = tp_agb_data['star_age']
three_m_log_L = tp_agb_data['log_L']
three_m_log_LHe = tp_agb_data['log_LHe']

three_m_surface_c12 = tp_agb_data['surface_c12']
three_m_surface_n14 = tp_agb_data['surface_n14'] 
three_m_surface_o16 = tp_agb_data['surface_o16']
```

```python
## 1 result here
plt.title('3$M_{\odot}$ Model During TP-AGB Phase')

plt.plot(three_m_star_age_years/1e9,
         three_m_log_L,label=r'$L$')

plt.plot(three_m_star_age_years/1e9,
         three_m_log_LHe,label=r'$L_{\rm{He}}$')


plt.xlim(1.225,1.227)

plt.legend()
plt.xlabel(r'$\rm{Age \ (Gyr)}$')
plt.ylabel(r'$\rm{log}~L \ (L_{\odot})$')
```

```python
## 2 result here
plt.title('3$M_{\odot}$ Model During TP-AGB Phase')

plt.plot(three_m_star_age_years/1e9,
         three_m_surface_c12,label=r'$X_{surf.}(^{12}\rm{C})$')

plt.plot(three_m_star_age_years/1e9,
         three_m_surface_n14,label=r'$X_{surf.}(^{14}\rm{N})$')

plt.plot(three_m_star_age_years/1e9,
         three_m_surface_o16,label=r'$X_{surf.}(^{16}\rm{O})$')


plt.xlim(1.225,1.227)


plt.legend()
plt.xlabel(r'$\rm{Age \ (Gyr)}$')
plt.ylabel(r'$\rm{Surface \ Mass \ Fraction}$')
```

### **With your larger group, try to answer the following**:

Using these plots answer the following:

> How many pulses are covered over this range?

11 pulses.

> What is their qualitative properties over time?

The amplitude of the pulse, shown via the helium shell luminosity grows over time.

> What is the approximate time between pulses, the _interpulse period_?

About total duration of 1.5Myr, for an average interpulse period of 0.13Myr, consistent with Figure 11.3.

> Hint: Compare with Pols [Figure 11.3/11.4](https://www.ucolick.org/~woosley/ay112-14/texts/pols11.pdf)


> Does this model experience a third dredge up event during this time? If so, how can we tell?

Yes, at about the 7th pulse the 3DU occurs and brings the now extinguished H-shell ash to the surface, in this case carbon and oxygen. 

> Does this model show evidence for hot bottom burning (HBB)?

No, its initial mass is too low to build a hot enough temperature to burn the bottom of the H-envelope via the CNO cycle. In a more maassive AGB star, this can occur and the evidence would be the conversion of C12 from the intershell region being proceed to N14 that is dredged to the surface. This thus prevents more massive AGB stars from becoming C (carbon) stars observationally.
